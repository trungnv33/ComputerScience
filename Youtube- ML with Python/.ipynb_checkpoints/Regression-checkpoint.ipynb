```python
import quandl as qd
import pandas as pd
import math
from sklearn import preprocessing, cross_validation, svm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import datetime
import matplotlib.pyplot as plt
from matplotlib import style
import pickle
style.use('ggplot')
```

```python
df = qd.get("WIKI/GOOGL")
```

# LÀM SẠCH DỮ LIỆU

```python
df.head()
```

```python
df=  df[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]
```

```python
df['HL_PCT'] = (df['Adj. High']- df['Adj. Close'])/ df['Adj. Close']*100.0
```

```python
df['PCT_change'] = (df['Adj. Close']- df['Adj. Open'])/ df['Adj. Open']* 100.0
```

```python
df = df[['Adj. Close','HL_PCT','PCT_change','Adj. Volume']]
df.head()
```

```python
forecast_col  = 'Adj. Close'
df[forecast_col]
```

```python
df.fillna(-99999,inplace = True)
```

```python
df
```

```python
forecast_out = int(math.ceil(0.01*len(df)))
forecast_out # số hàng đẩy ngược lên trên 
```

# DỊCH CHUYỂN ROW LÊN SỐ FORECAST_OUT ĐỂ THỪA RA CÁC GIÁ TRỊ NAN TRONG LABEL ĐỂ DỰ ĐOÁN SAU NÀY

```python
df['label'] = df[forecast_col].shift(-forecast_out)
df.tail()
```

```python
df.head()
```

```python
df.dropna(inplace = True)
df.tail()
```

* Đặt biến đầu vào và ra


```python
X = np.array(df.drop(['label'],1))
X = preprocessing.scale(X)
X_lately = X[-forecast_out:]
X = X[:-forecast_out]
df.dropna(inplace=  True)
y = np.array(df['label'])
y = y[:-forecast_out]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
clf = LinearRegression()
clf.fit(X_train,y_train)
accuracy = clf.score(X_test,y_test)
```

```python
print(accuracy)
```

* Bắt đầu quá trình dự đoán:

```python
forecast_set = clf.predict(X_lately)
print(forecast_set, accuracy, forecast_out)
```

# VISUALIZE THE PREDICTIONS

```python
days = range(1,len(forecast_set)+1)
plt.plot(days, forecast_set)
plt.xlabel('Days')
plt.ylabel('Predicted Price (%$)')
```
